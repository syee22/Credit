### ‚úÖ Ïù¥ÏÉÅÏπò Ï†úÍ±∞

### Ïù¥ÏÉÅÏπòÏùò Ï°∞Í±¥ : Q3 + IQR * 3 Î•º Ï¥àÍ≥ºÌïòÎ©¥ Ïù¥ÏÉÅÏπò, Q1 - IQR * 3 ÎØ∏ÎßåÏù¥Î©¥ Ïù¥ÏÉÅÏπò
- 1.5Î°ú Ï°∞Í±¥ÏùÑ ÌñàÏùÑ Í≤ΩÏö∞ ÏïΩ 7000Í∞úÏùò Î†àÏΩîÎìúÍ∞Ä ÏÇ≠Ï†úÎêòÏñ¥ Îç∞Ïù¥ÌÑ∞ ÏàòÍ∞Ä ÏßÄÎÇòÏπòÍ≤å Ï†ÅÏñ¥Ïßê
- ÏïÑÏù¥ Ïàò, ÏÜåÎìù, Í∞ÄÏ°±ÌÅ¨Í∏∞ Îì±ÏùÄ ÏïÑÏ£º Í∑πÎã®Ï†ÅÏù∏ Ïù¥ÏÉÅÏπòÎßå ÏÇ≠Ï†úÌïòÎ©¥ ÎêúÎã§Í≥† ÌåêÎã®Ìï®  **ÏïÑÏù¥Ïàò Í∞ÄÏ°±ÌÅ¨Í∏∞Í¥ÄÍ≥ÑÌôïÏù∏**

```python
# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†Ñ train ÌòïÌÉú ÌôïÏù∏
out_before = train.shape
out_before
```

```python
# Ïù¥ÏÉÅÏπò Íµ¨Ìï¥Ï£ºÎäî Ìï®Ïàò
def IQR_rule(val_list) : 
    # IQR Í≥ÑÏÇ∞
    Q1 = np.quantile(val_list, 0.25)
    Q3 = np.quantile(val_list, 0.75)
    IQR = Q3 -Q1
    
    # IQR RuleÏùÑ ÏúÑÎ∞∞ Í≤∞Í≥ºÎ•º bool listÎ°ú ÏÉùÏÑ± ÌõÑÏóê Î∞òÌôò
    # Í≤∞Í≥ºÍ∞íÏù¥ TrueÎ©¥ Ï†ïÏÉÅÎç∞Ïù¥ÌÑ∞, Í≤∞Í≥ºÍ∞íÏù¥ FalseÎ©¥ Ïù¥ÏÉÅÏπò
    not_outlier_condition = (Q3 + 3 * IQR > val_list) & (Q1 - 3 * IQR < val_list)
    return not_outlier_condition
```

```python
# numerical_feats Ï§ëÏóêÏÑú Í≤∞Í≥ºÌîºÏ≤ò(credit)ÏôÄ Î≥ÄÏàòÏùò Í∞ØÏàòÍ∞Ä 2Í∞ú Ïù¥ÌïòÏù∏ Í≤ÉÏùÄ Ï†úÏô∏
columns_list = ['child_num', 'income_total', 'DAYS_BIRTH', 'DAYS_EMPLOYED', 'family_size', 'begin_month' ]

df = pd.DataFrame()
```

```python
# Ïù¥ÏÉÅÏπòÍ∞Ä ÏûàÏùåÎ©¥ Îç∞Ïù¥ÌÑ∞ÌîÑÎ†àÏûÑÏóê FalseÎ°ú ÌëúÏãú
for i in columns_list:
    conditions = IQR_rule(train[i]) # apply() :ÌååÎùºÎØ∏ÌÑ∞Î°ú Ï†ÑÎã¨ÎêòÎäî Ìï®ÏàòÎ•º Ïó¥Î∞©Ìñ•ÏúºÎ°ú Ï†ÅÏö©ÏãúÏºúÏ£ºÎäî Ìï®Ïàò
    df = pd.concat([df, conditions], axis=1)
df
```

```python
# trainÏóêÏÑú Ïù¥ÏÉÅÏπò Îç∞Ïù¥ÌÑ∞ Ï†úÍ±∞
total_conditions = df.sum(axis=1) == 6
train = train.loc[total_conditions]
```

```python
# Ïù¥ÏÉÅÏπò Ï†úÍ±∞ ÌõÑ train ÌòïÌÉú ÌôïÏù∏
out_after = train.shape
print('Ïù¥ÏÉÅÏπò Ï†úÍ±∞ Ï†Ñ',out_before,'Ïù¥ÏÉÅÏπò Ï†úÍ±∞ ÌõÑ',out_after)
print('Ï¥ù ',out_before[0]-out_after[0],'Í∞úÏùò Ìñâ Ï†úÍ±∞')
```

### 'DAYS_BIRTH', 'begin_month','DAYS_EMPLOYED' Ï†àÎåÄÍ∞íÏúºÎ°ú Î≥ÄÌôò

```python
train['DAYS_EMPLOYED'] = train['DAYS_EMPLOYED'].map(lambda x: 0 if x > 0 else x)
test['DAYS_EMPLOYED'] = test['DAYS_EMPLOYED'].map(lambda x: 0 if x > 0 else x)
```

```python
features = ['DAYS_BIRTH', 'begin_month','DAYS_EMPLOYED']
for f in features:
    train[f] = np.abs(train[f])
    test[f] = np.abs(test[f])
```

### ‚úÖ ÌååÏÉùÎ≥ÄÏàò ÏÉùÏÑ±

```python
for df in [train,test]:
    # ÎÇòÏù¥ 
    df['Age'] = df['DAYS_BIRTH'] // 365
    
    # EMPLOYED(Í∑ºÏÜçÏó∞Ïàò)
    df['EMPLOYED'] = df['DAYS_EMPLOYED'] // 365
    
    # ability: ÏÜåÎìù/(ÏÇ¥ÏïÑÏò® ÏùºÏàò+ Í∑ºÎ¨¥ÏùºÏàò)
    df['ability'] = df['income_total'] / (df['DAYS_BIRTH'] + df['DAYS_EMPLOYED'])
    
    # income_mean: ÏÜåÎìù/ Í∞ÄÏ°± Ïàò 
    df['income_mean'] = df['income_total']/ df['family_size']
    
    # contact_pos: Ïó∞ÎùΩÍ∞ÄÎä•ÏÑ±
    df['contact_pos'] = df['FLAG_MOBIL'] + df['work_phone'] + df['phone'] + df['email']
    
    # childcare_bur: ÏñëÏú°Î∂ÄÎã¥ÎèÑ
    df['childcare_bur'] = df['child_num'] / df['income_total'] * 1000000 
    
    #ID ÏÉùÏÑ±: Í∞Å Ïª¨ÎüºÏùò Í∞íÎì§ÏùÑ ÎçîÌï¥ÏÑú Í≥†Ïú†Ìïú ÏÇ¨ÎûåÏùÑ ÌååÏïÖ + begin_month countÌï¥ÏÑú Ïπ¥ÎìúÍ∞ØÏàò Ïª¨Îüº ÏÉùÏÑ± 
    df['ID'] = \
    df['child_num'].astype(str) + '_' + df['income_total'].astype(str) + '_' +\
    df['DAYS_BIRTH'].astype(str) + '_' + df['DAYS_EMPLOYED'].astype(str) + '_' +\
    df['work_phone'].astype(str) + '_' + df['phone'].astype(str) + '_' +\
    df['email'].astype(str) + '_' + df['family_size'].astype(str) + '_' +\
    df['gender'].astype(str) + '_' + df['car'].astype(str) + '_' +\
    df['reality'].astype(str) + '_' + df['income_type'].astype(str) + '_' +\
    df['edu_type'].astype(str) + '_' + df['family_type'].astype(str) + '_' +\
    df['house_type'].astype(str) + '_' + df['occyp_type'].astype(str)
```

```python

folds=[]
for train_idx, valid_idx in skfold.split(X, y):
        folds.append((train_idx, valid_idx))

cat_pred = np.zeros((X.shape[0], n_class))
cat_pred_test = np.zeros((X_test.shape[0], n_class))
# cat_cols = categorical_feats
cat_cols = ['income_type', 'edu_type', 'family_type', 'house_type', 'occyp_type']

for fold in range(n_fold):
    
    print(f'\n----------------- Fold {fold} -----------------\n')
    train_idx, valid_idx = folds[fold]
    X_train, X_valid, y_train, y_valid = X.iloc[train_idx], X.iloc[valid_idx], y[train_idx], y[valid_idx]

#     train_data = Pool(data=X_train, label=y_train, cat_features=cat_cols)
#     valid_data = Pool(data=X_valid, label=y_valid, cat_features=cat_cols)
    train_data = Pool(data=X_train, label=y_train)
    valid_data = Pool(data=X_valid, label=y_valid)

    model_cat = CatBoostClassifier()
    model_cat.fit(train_data, eval_set=valid_data, use_best_model=True, early_stopping_rounds=100, verbose=100)

    cat_pred[valid_idx] = model_cat.predict_proba(X_valid)
    cat_pred_test += model_cat.predict_proba(X_test) / n_fold
    print(f'CV Log Loss Score: {log_loss(y_valid, cat_pred[valid_idx]):.6f}')

print(f'\tLog Loss: {log_loss(y, cat_pred):.6f}')
```
