```python
import pandas as pd
import numpy as np
```

```python
f4 = pd.read_csv('./scoring/모델별 점수_table_파라미터설정.csv',index_col=0)
f4.name = '라벨인코딩**minmax**로그변환++up'
f4
```

```python
f1 = pd.read_csv('./scoring/모델별 점수_table.csv',index_col=0)
f1.name = '라벨인코딩**minmax**로그변환'
f1
```

```python
f2 = pd.read_csv('./scoring/one-hot_encoding_모델별 점수_table.csv',index_col=0)
f2.name = '원핫인코딩**로그변환'
f2
```

```python
f3 = pd.read_csv('./scoring/log와 standard_label적용.csv',index_col=0)
f3.name = '정규화**로그변환**라벨인코딩'
f3
```

```python
f_0 = pd.read_csv('./scoring/적용최소화 점수_table_파라미터설정.csv',index_col=0)
f_0.name = '라벨인코딩'
f_0
```

```python
f = pd.read_csv('./scoring/모델별 점수_table_파라미터설정_down.csv',index_col=0)
f.name = '라벨인코딩**minmax**로그변환++down'
f
```

```python
model_list = [f_0,f,f1,f4,f2,f3]
name_list =[f_0.name,f.name,f1.name,f4.name,f2.name,f3.name]
```

# ⭕GBM

```python
GBM_scores = pd.DataFrame()
for df in model_list:
    GBM_scores = pd.concat([GBM_scores,df.iloc[0]],axis=1)
GBM_scores.columns = name_list
```

# ⭕ RandomForest

```python
RF_scores = pd.DataFrame()
for df in model_list:
    RF_scores = pd.concat([RF_scores,df.iloc[1]],axis=1)
RF_scores.columns=name_list
RF_scores.head()
```

# ⭕ DecisionTree

```python
DT_scores = pd.DataFrame()
for df in model_list:
    DT_scores = pd.concat([DT_scores,df.iloc[2]],axis=1)
DT_scores.columns=name_list
DT_scores.head()
```

# ⭕ LGBM

```python
LGBM_scores = pd.DataFrame()
for df in model_list:
    LGBM_scores = pd.concat([LGBM_scores,df.iloc[3]],axis=1)
LGBM_scores.columns= name_list
LGBM_scores.head()
```

# ⭕ XGB

```python
XGBoost_scores = pd.DataFrame()
for df in model_list:
    XGBoost_scores = pd.concat([XGBoost_scores,df.iloc[4]],axis=1)
XGBoost_scores.columns=name_list
XGBoost_scores.head()
```

# ⭕ SOFTVOTING

```python
Voting_scores = pd.DataFrame()
for df in model_list:
    Voting_scores = pd.concat([Voting_scores,df.iloc[5]],axis=1)
Voting_scores.columns=name_list
Voting_scores.head()
```

# 🟢 MODELING SCORE

```python
model_score = pd.concat([GBM_scores,RF_scores,DT_scores,LGBM_scores,XGBoost_scores,Voting_scores],axis=1).T
model_score
```

```python
model_score.to_csv('Modeling_점수.csv')
```
